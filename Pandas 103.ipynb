# PROBLEM 1: Display 'Manager' and 'Grand Total Sales' done by the Sales Agents reporting these managers

```python
# Manager and coresponding sales agent 

sales_combo = pd.merge(sales_teams,sales_pipeline,how = 'inner',on = 'sales_agent')
```

```python
sales_combo.head()
```

```python
temp1 = sales_combo[sales_combo['deal_stage'] == "Won"]
temp1 = temp1[['manager','sales_agent','close_value']]
```

```python
temp2 = temp1.groupby('sales_agent').sum().sort_values(by = 'close_value',ascending = False)
temp2 = pd.DataFrame(temp2)
```

```python
temp2.reset_index(inplace = True)
```

```python
temp2.head()
```

```python
pd.merge(sales_teams,temp2, how = 'inner',on = 'sales_agent').drop('regional_office',axis = 1)
```
